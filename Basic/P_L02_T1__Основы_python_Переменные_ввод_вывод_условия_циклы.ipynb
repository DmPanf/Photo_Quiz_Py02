## **(МОДУЛЬ 3) ``bornday.py``**
- **https://raw.githubusercontent.com/dnp34/Python_L02/main/bornday.py**

```python
# Спросить у пользователя год рождения А.С. Пушкина
# Если пользователь ввел верный год спросить у него день рождения
# Если день рождения введен верно вывести 'Верно'
# Если день рождения введен неверно вывести 'Неверный день рождения'
# Если неверно введен год, то сразу вывести 'Неверный год
# рождения', а день рождения не спрашивать

Year = 1799
Date = ['6.06', '6/06', '6.6', '6/6', '6-06', '6-6', '6']

for i in range(3):
  X = int(input(f'Введите год рождения А.С Пушкина [осталось попыток - {3-i}]: '))
  if X == Year:
    Y = input(f'.... введите день рождения: ')
    if Y in Date:
      print('Верно')
    else:
      print('Неверный день рождения!')
    break
  else:
    print('Неверный год рождения..')
```

## **(МОДУЛЬ 5) ``borndayforewer.py``**
- **https://raw.githubusercontent.com/dnp34/Python_L02/main/borndayforewer.py**

```python
# Используя условные операторы и цикл while: Просим пользователя ввести год рождения А.С. Пушкина 
# до тех пор пока он не ввел правильный год, после этого спрашиваем день рождения до тех пор, 
# пока он не ввел верный день. Выводим в терминал 'Верно' и выходим из программы

Year = 1799
Date = ['6.06', '6/06', '6.6', '6/6', '6-06', '6-6', '6']

X = int(input('\nВведите год рождения А.С Пушкина: '))

while X != Year:
  X = int(input('Неверно! Введите год рождения А.С Пушкина: '))
  if X == Year:
    Y = input('.... введите день рождения: ')
    while Y not in Date:
      Y = input('.... Вы ошибаетесь. Введите день рождения: ')
      if Y in Date:
        print('Верно')
```

## **(МОДУЛЬ 6) ``victory.py``**
- **https://raw.githubusercontent.com/dnp34/Python_L02/main/victory.py**

```python
# Программа викторина: Выбрать минимум 5 известных людей и узнать их год рождения. 
# Программа должна по очереди спрашивать у пользователя год рождения знаменитости. 
# После того как пользователь ввел все ответы, программа считает и выводит на экран:
# - количество правильных ответов
# - количество ошибок
# - процент правильных ответов (можно посчитать как ПРАВИЛЬНЫЕ ОТВЕТЫ*100/ВСЕГО ВОПРОСОВ)
# - процент неправильных ответов 
# После вывода информации программа спрашивает пользователя хочет ли он начать игру сначала, 
# если да - то повторяем все сначала, если ответ нет - выходим из программы

People = {'А.С. Пушкина': 1799, 'Джека Лондона': 1876, 'М.В. Ломоносова': 1711, 'Д.И. Менделеева': 1834, 'С.В. Ковалевской': 1850, 'Исаака Ньютона': 1643, 'Алана Тьюринга': 1936, 'С.П. Королева': 1907}
YesNo = True

def Game(): # Функция Игры Викторина
  print('\nСТАРТ!!!')
  good_num = 0 # Обнуление счетчика правильных ответов
  All = len(People) # Общее число вопросов

  for key, val in People.items(): # Перебор пар Словаря (ключ и значение)
    Z = int(input(f'🔘 назовите год рождения {key}: ')) # Присвоение введенного года в переменную Z
    if Z == val: # Сравнение введенного года со значением в Словаре
      good_num += 1 # При угадывании увеличение счетчика правильных ответов

  Mist = All - good_num # Вычисление количества ошибок
  print(f'\n✅ количество правильных ответов: {good_num} [{round((good_num * 100) / All, 0)} %] из {All}') # Вывод количества и %% правильных ответов
  print(f'❌ количество ошибок: {Mist} [{round((Mist * 100) / All, 0)} %] из {All}') # Вывод количества и %% ошибок
  Answer = input('\nНе желаете начать игру сначала? [да][нет]: ') # Запрос на повторную игру
  if Answer.lower() == 'да' or Answer.lower() == 'yes': # Проверка ответа (возможны рус. и англ. яз. в любом написании строчными или заглавными буквами)
    YesNo = True # Если 'да'/'yes', то продолжить игру
  else:
    YesNo = False # В иных случаях - останов
  return YesNo

print('\n♻️ Начинаем викторину.\nЗадача: правильно вводить год рождения знаменитости!\n')
while YesNo: # Бесконечный цикл, пока будет вводится 'да'/'yes' на повтор игры
  YesNo = Game() # Вызов Функции, которая возвращает логический тип
```
